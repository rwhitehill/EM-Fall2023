```python
import numpy as np
import matplotlib.pyplot as plt
plot_params = {
    'text.latex.preamble' : r'\usepackage{amsmath}',
    'text.usetex'         : True,
    'font.family'         : 'sans-serif',
    'font.sans-serif'     : ['Helvetica']   
}
plt.rcParams.update(plot_params)
import sympy as sp
```

# Problem 2

```python
x1 = np.linspace(0,1)
x2 = np.linspace(1,5)

nrows=1;ncols=3
fig,ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=(7*ncols,6*nrows))

ax[0].plot(x1,np.zeros(len(x1)),'r-')
ax[0].plot(x2,1/x2**2,'r-')
ax[0].plot([1,1],[0,1],'r-')
ax[0].text(x=0.95,y=0.95,s=r'$\epsilon_0 E_r/\sigma$',size=30,ha='right',va='top',transform=ax[0].transAxes)

ax[1].plot(x1,x1,'r-')
ax[1].plot(x2,1/x2**2,'r-')
ax[1].text(x=0.95,y=0.95,s=r'$3 \epsilon_0 E_r/a \rho_0$',size=30,ha='right',va='top',transform=ax[1].transAxes)

for n in [-2,2]:
    ax[2].plot(x1,x1**(n+1),color='r',ls='-' if n == -2 else '-.',label=r'$n = %d$'%n)
    ax[2].plot(x2,1/x2**2,color='r',ls='-' if n == -2 else '-.')
ax[2].text(x=0.95,y=0.95,s=r'$(n+3) \epsilon_0 E_r/a^{n+1} \rho_0$',size=30,ha='right',va='top',transform=ax[2].transAxes)
ax[2].legend(fontsize=20,loc='center right')
ax[2].set_ylim(-0.3,5)

for i in range(ncols):
    ax[i].tick_params(axis='both',which='major',direction='in',labelsize=20)
    ax[i].set_xlabel(r'$r/a$',size=30)
    # ax[i].set_ylim(-0.1,1.1)

plt.tight_layout()
plt.show()
fig.savefig('prob2.pdf',bbox_inches='tight')
```

# Problem 3

```python
r,alpha = sp.symbols(r'r \alpha',real=True,positive=True)
expr = sp.exp(-alpha*r)/r*(1 + alpha*r/2)
expr = (r**2*expr.diff(r)).diff(r)/r**2
expr.simplify()
```

```python
x = np.linspace(0,1)

nrows=1;ncols=3
fig,ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=(7*ncols,6*nrows))

ax[0].plot(x,np.ones(len(x)),'r-')
ax[0].set_xlabel(r'$z/d$',size=30)
ax[0].text(x=0.95,y=0.95,s=r'$(2\epsilon_0 A/Q^2)u_E$',size=30,ha='right',va='top',transform=ax[0].transAxes)

for xi in [0.5,0.8]:
    ax[1].plot(x,1/((1-xi)*x+xi)**4,color='r',ls='-' if xi==0.5 else '-.',label=r'$a = %.1f b$'%xi)
ax[1].legend(loc='center right',fontsize=20,frameon=False)
ax[1].set_xlabel(r'$(r-a)/(b-a)$',size=30)
ax[1].text(x=0.95,y=0.95,s=r'$(32 \pi^2 \epsilon_0 b^4/Q^2) u_E$',size=30,ha='right',va='top',transform=ax[1].transAxes)

for xi in [0.5,0.8]:
    ax[2].plot(x,1/((1-xi)*x+xi)**2,color='r',ls='-' if xi==0.5 else '-.',label=r'$a = %.1f b$'%xi)
ax[2].legend(loc='center right',fontsize=20,frameon=False)
ax[2].set_xlabel(r'$(\rho-a)/(b-a)$',size=30)
ax[2].text(x=0.95,y=0.95,s=r'$(8 \pi^2 \epsilon_0 b^2/\lambda^2) u_E$',size=30,ha='right',va='top',transform=ax[2].transAxes)

for i in range(ncols):
    ax[i].tick_params(axis='both',which='major',direction='in',labelsize=20)

plt.tight_layout()
plt.show()
fig.savefig('prob5.pdf',bbox_inches='tight')
```
