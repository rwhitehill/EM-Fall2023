```python
x = np.linspace(1,10)

fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(7*2,5))
ax[0].plot(x,1/x**3,color='k')
ax[0].set_xlabel(r'$r/a$',size=30)
ax[0].set_ylabel(r'$-(4 \pi \epsilon_0 a / q) \Phi$',size=30)
ax[0].tick_params(axis='both',which='major',direction='in',labelsize=20)

ax[1].plot(x,np.ones(len(x)),color='k',label=r'$\rm multipole$')
ax[1].plot(x,2*(x**2 - x**3/np.sqrt(x**2+1)),label=r'$\rm exact$')
ax[1].set_xlabel(r'$r/a$',size=30)
ax[1].set_ylabel(r'$-(4 \pi \epsilon_0 a / q) (r/a)^3 \Phi$',size=30)
ax[1].tick_params(axis='both',which='major',direction='in',labelsize=20)
ax[1].legend(fontsize=25,loc='lower right',frameon=False)

plt.tight_layout()
plt.show()
fig.savefig('prob2.pdf',bbox_inches='tight')
```
